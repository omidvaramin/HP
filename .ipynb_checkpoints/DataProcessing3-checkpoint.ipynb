### remove null posts

```python
#keep only characters and space
def keep_chars(title):
    return re.sub(r'[^A-Za-z ]+', '', title)
```

```python
def keep_only_one_space(title):
    return ' '.join(title.split())
```

```python
#remove null posts
print(len(MergedData))
MergedData = MergedData[~MergedData['post_text'].isnull()]
print(len(MergedData))
```

### Removing outliers in terms of favourite_count for each author

```python
MergedData.boxplot(by = 'author_name'
                    , column =['favorite_count']
                    , grid = False
                    ,rot = 45
                    ,fontsize='large'
                   ,figsize=(20,10)
                  ,showfliers=False)
```

```python
#get the unique number of news meida and their posts

def q25(x):
    return x.quantile(0.25)

def q7(x):
    return x.quantile(0.7)

def q95(x):
    return x.quantile(0.95)

def q9(x):
    return x.quantile(0.9)

def q8(x):
    return x.quantile(0.8)

def q1(x):
    return x.quantile(0.05)

f = {'favorite_count': ['count','min',q1,q25,'mean','median',q7,q9,q8,'max']}


res_quantile = MergedData[['user_name','favorite_count']].groupby(['user_name']).agg(f).reset_index()
```

```python
res_quantile
```

```python
uppder_bound_threshold = 'q9'
res_upper_treshold_dict = {}
for i in range(len(res_quantile)):
    user_name =  res_quantile.iloc[i]['user_name'][0]
    
    threshold = res_quantile.iloc[i]['favorite_count'][uppder_bound_threshold]
    res_upper_treshold_dict[user_name] = threshold
```

```python
res_upper_treshold_dict
```

```python
Clean_MergedData = MergedData.copy()
Clean_MergedData[f'{uppder_bound_threshold}_outlier'] = 0
Clean_MergedData[f'{uppder_bound_threshold}_favorite_count'] = 0
for i in range(len(MergedData)):
    user_name = Clean_MergedData.iloc[i]['user_name']
    upper_threshold = res_upper_treshold_dict[user_name]
    favorite_count = Clean_MergedData.iloc[i]['favorite_count']
    if favorite_count >= upper_threshold:
        Clean_MergedData.iloc[i, Clean_MergedData.columns.get_loc(f'{uppder_bound_threshold}_outlier')] = 1
        Clean_MergedData.iloc[i, Clean_MergedData.columns.get_loc(f'{uppder_bound_threshold}_favorite_count')] = upper_threshold
```

```python
Clean_MergedData = MergedData[MergedData[f'{uppder_bound_threshold}_outlier'] == 0]
print(len(MergedData))
print(len(Clean_MergedData))
```

```python
Clean_MergedData.boxplot(by = 'author_name'
                    ,column =['favorite_count']
                    ,grid = False
                    ,rot = 80
                    ,fontsize=14
                   ,figsize=(20,7)
                   ,showfliers=True)
plt.savefig('favorite_count.png', dpi = 300)
```

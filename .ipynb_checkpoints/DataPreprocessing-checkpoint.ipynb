```python
def get_follow_cnt(username):
    return df_user_follow_cnt_median[df_user_follow_cnt_median['user_name'] == username]['users_followers_count'].iloc[0]

get_follow_cnt('USA TODAY')
```

```python

alldatastats1['users_followers_count_median'] = 0
counter = 0 
for idx, row in alldatastats1.iterrows():
    counter+=1
    if counter % 1000 == 0:
        print(counter*100/len(alldatastats1))
    followers_cnt = get_follow_cnt(row['user_name'])
    #print(row['user_name'], '   ', idx,'   ',followers_cnt)
    alldatastats1.loc[idx,'users_followers_count_median'] = followers_cnt
```

```python
#Load all ids
def read_data(fps, word2id=None, y_len=1, use_target_description=False, use_image=False, delete_irregularities=False):
    ids = []
    post_texts = []
    post_text_lens = []
    truth_means = []
    truth_classes = []
    id2truth_class = {}
    id2truth_mean = {}
    target_descriptions = []
    target_description_lens = []
    image_features = []    
    num = 0
    target_paragraphs = []
    target_first_par = []
    for fp in fps:
        if use_image:
            with open(os.path.join(fp, "id2imageidx.json"), "r") as fin:
                id2imageidx = json.load(fin)
            #all_image_features = hickle.load(os.path.join(fp, "image_features.hkl"))
            all_image_features = pickle.load(os.path.join(fp, "image_features.hkl"))
        if y_len:
            with open(os.path.join(fp, 'truth.jsonl'), 'rb') as fin:
                for each_line in fin:
                    each_item = json.loads(each_line.decode('utf-8'))
                    if delete_irregularities:
                        if each_item["truthClass"] == "clickbait" and float(each_item["truthMean"]) < 0.5 or each_item["truthClass"] != "clickbait" and float(each_item["truthMean"]) > 0.5:
                            continue
                    if y_len == 4:
                        each_label = [0, 0, 0, 0]
                        #for each_key, each_value in Counter(each_item["truthJudgments"]).iteritems():
                        for each_key, each_value in Counter(each_item["truthJudgments"]).items():
                            each_label[int(each_key//0.3)] = float(each_value)/5
                        id2truth_class[each_item["id"]] = each_label
                        if each_item["truthClass"] != "clickbait":
                            assert each_label[0]+each_label[1] > each_label[2]+each_label[3]
                        else:
                            assert each_label[0]+each_label[1] < each_label[2]+each_label[3]
                    if y_len == 2:
                        if each_item["truthClass"] == "clickbait":
                            id2truth_class[each_item["id"]] = [1, 0]
                        else:
                            id2truth_class[each_item["id"]] = [0, 1]
                    if y_len == 1:
                        if each_item["truthClass"] == "clickbait":
                            id2truth_class[each_item["id"]] = [1]
                        else:
                            id2truth_class[each_item["id"]] = [0]
                    id2truth_mean[each_item["id"]] = [float(each_item["truthMean"])]
        
        with open(os.path.join(fp, 'instances.jsonl'), 'rb') as fin:
            for each_line in fin:
                each_item = json.loads(each_line.decode('utf-8'))
                if each_item["id"] not in id2truth_class and y_len:
                    num += 1
                    continue
                ids.append(each_item["id"])
                each_post_text = " ".join(each_item["postText"])
                each_target_description = each_item["targetTitle"]
                paragraphs = each_item["targetParagraphs"]
                if y_len:
                    truth_means.append(id2truth_mean[each_item["id"]])
                    truth_classes.append(id2truth_class[each_item["id"]])
                if word2id:
                    if (each_post_text+" ").isspace():
                        #the id of <unk>
                        post_texts.append([0])
                        post_text_lens.append(1)
                    else:
                        each_post_tokens = tokenise(each_post_text)
                        post_texts.append([word2id.get(each_token, 1) for each_token in each_post_tokens])
                        post_text_lens.append(len(each_post_tokens))
                else:
                    post_texts.append([each_post_text])
                    
                
                merged_pars = ''
                first_par = ''
                for i in range(len(paragraphs)):
                    merged_pars = merged_pars + ' ' + paragraphs[i]
                    if i == 0:
                        first_par = paragraphs[i]
            
                target_paragraphs.append(merged_pars)
                target_first_par.append(first_par)
                
                if use_target_description:
                    if word2id:
                        if (each_target_description+" ").isspace():
                            target_descriptions.append([0])
                            target_description_lens.append(1)
                        else:
                            each_target_description_tokens = tokenise(each_target_description)
                            target_descriptions.append([word2id.get(each_token, 1) for each_token in each_target_description_tokens])
                            target_description_lens.append(len(each_target_description_tokens))
                    else:
                        target_descriptions.append([each_target_description])
                else:
                    target_descriptions.append([])
                    target_description_lens.append(0)
                if use_image:
                    image_features.append(all_image_features[id2imageidx[each_item["id"]]].flatten())
                else:
                    image_features.append([])
    print ("Deleted number of items: " + str(num))
    return ids, post_texts, truth_classes, post_text_lens, truth_means, target_descriptions, target_description_lens, image_features,target_paragraphs, target_first_par
```

```python
ids, post_texts, truth_classes, post_text_lens, truth_means, target_descriptions \
, target_description_lens, image_features, target_paragraphs, target_first_pars =\
                    read_data(word2id=None, fps=['dataCrawling/clickbait17-unlabeled-170429'
                                                 ,'dataCrawling/clickbait17-validation-170630'
                                                 ,'dataCrawling/clickbait17-train-170331'],                               
                              y_len=None, use_target_description=True, use_image=False)
```

```python
idsl, post_textsl, truth_classesl, post_text_lensl, truth_meansl, target_descriptionsl \
, target_description_lensl, image_featuresl, target_paragraphsl, target_first_parsl =\
                    read_data(word2id=None, fps=['dataCrawling/clickbait17-validation-170630'
                                                 ,'dataCrawling/clickbait17-train-170331'],                               
                              y_len=1, use_target_description=True, use_image=False)

id2truth_meansl = {}
for i in range(len(idsl)):
    id2truth_meansl[idsl[i]] = truth_meansl[i]
```

```python
#Merging all data in one dataframe
MergedDatalist = []

counter = 0


for i in range(len(ids)):
    counter+=1
    if counter % 1000 == 0:
        print(counter*100/len(ids))
    
    #print(i)
    ID = ids[i]
    ID = int(ID)
    #print(ID)
    post_text = post_texts[i][0]
    #truth_class = truth_classes[i][0]
    #truth_mean = truth_means[i][0]
    target_description = target_descriptions[i][0]
    target_pars = target_paragraphs[i]
    first_par = target_first_pars[i]
        
    
    res = alldatastats.loc[alldatastats['id'] == ID]
    if(len(res) == 0):
        continue
        
        
    
    favorite_count = res['favorite_count'].iloc[0]
    retweet_count = res['retweet_count'].iloc[0]
    created_at = res['created_at'].iloc[0]
    favorited = res['favorited'].iloc[0]
    retweeted = res['retweeted'].iloc[0]
    users_followers_count = res['users_followers_count'].iloc[0]
    user_name = res['user_name'].iloc[0]
    user_description = res['user_description'].iloc[0]
    user_url = res['user_url'].iloc[0]
    author_name = res['author_name'].iloc[0]
    author_friends_count = res['author_friends_count'].iloc[0]
    users_followers_count_median = res['users_followers_count_median'].iloc[0]
    outlier_flagged = res['outlier_flagged'].iloc[0]
    fav_count_over_followers = res['fav_count_over_followers'].iloc[0]
    
    if ids[i] in id2truth_meansl:
        truth_mean = id2truth_meansl[ids[i]][0]
        if truth_mean > 0.5:
            truth_class = 1
        else:
            truth_class = 0
    else:
        truth_mean = None
        truth_class = None
    
    MergedDatalist.append([ID,
                           post_text,
                           truth_class,
                           truth_mean,
                           target_description,
                           target_pars,
                           first_par,
                           favorite_count,
                           retweet_count,
                           created_at,
                           favorited,
                           retweeted,
                           users_followers_count,
                           user_name,
                           user_description,
                           user_url,
                           author_name,
                           author_friends_count,
                           users_followers_count_median,
                           outlier_flagged,
                           fav_count_over_followers
                          ])
```

### removing outliers


```python
#get the unique number of news meida and their posts

def q25(x):
    return x.quantile(0.25)

def q7(x):
    return x.quantile(0.7)

def q95(x):
    return x.quantile(0.95)

def q9(x):
    return x.quantile(0.9)

def q8(x):
    return x.quantile(0.8)

def q1(x):
    return x.quantile(0.05)

f = {'favorite_count': ['count','min',q1,q25,'mean','median',q7,q9,q8,'max']}


res_quantile = alldatastats[['user_name','favorite_count']].groupby(['user_name']).agg(f).reset_index()
```

```python
alldatastats.head(1)
```

```python
res_quantile.iloc[0]['user_name'][0]
```

```python
res_q9_dict = {}
```

```python
for i in range(len(res_quantile)):
    user_name =  res_quantile.iloc[i]['user_name'][0]
    
    threshold = res_quantile.iloc[0]['favorite_count']['q7']
    res_q9_dict[user_name] = threshold
```

```python
alldatastats['q7outlier'] = 0
```

```python
for i in range(len(alldatastats)):
    user_name = alldatastats.iloc[i]['user_name']
    q9_threshold = res_q9_dict[user_name]
    favorite_count = alldatastats.iloc[i]['favorite_count']
#     print(i)
#     print(favorite_count)
#     print(q9_threshold)
    if favorite_count >= q9_threshold:
        alldatastats.iloc[i, alldatastats.columns.get_loc('q7outlier')] = 1
```
